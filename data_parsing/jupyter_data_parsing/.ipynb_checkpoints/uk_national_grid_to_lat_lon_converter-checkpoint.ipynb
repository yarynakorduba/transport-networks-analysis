```python
from convertbng.util import convert_bng, convert_lonlat
```

```python
with open("bristol_data_latest_version.json", "r") as file_with_stops:
    stops_data = file_with_stops.read()
stops_parsed_data = json.loads(stops_data)
```

```python
for stop in stops_parsed_data:
    lon_lat = convert_lonlat([int(stop["easting"])],[int(stop["northing"])]) #lat lon
    stop["lon"] = lon_lat[0][0]
    stop["lat"] = lon_lat[1][0]
```

```python
with open("../data/latest_bristol_BUS_stops_with_gps_positions.json", "w") as final_file:
    json.dump(stops_parsed_data, final_file)
```

```python
cleaned_parsed_data = []
for stop in stops_parsed_data:
   cleaned_parsed_data.append({
        'lon': stop["lon"],
        "lat": stop["lat"],
        "routes": stop["routes"],
        "label": stop["label"],
        "connections": stop["connections"],
        "stationType": stop["stationType"],
        "id": stop["id"]
    })
```

```python
print(cleaned_parsed_data)
```

```python
with open("../data/BRISTOL_LATEST_WITH_LONDON_MISTAKE.json", "w") as final_file:
    json.dump(cleaned_parsed_data, final_file)
```
